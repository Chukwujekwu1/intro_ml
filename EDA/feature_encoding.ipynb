Feature encoding is the act of changing categorical data to numerical value

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
# Import and read the csv file
df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
# Droping colums i do need/ useless columns
df.drop(columns = ["CustomerId","RowNumber","Surname"], axis=1,inplace=True)
df.info()
```

```python
df.isnull().sum()
```

```python
# Handeling missing values
n_df = df
n_df["Age"]  = n_df["Age"].fillna(n_df["Age"].mean())
```

```python
n_df.info()
```

<h3 style="font-weight:bold">1. Label Encoding</h3>

<p style="font-size:1.2rem">This type of encoding is only used for encoding y-target variabels <br> It works by assigning a number to every variable. e.g  Gender column has value of male and female the algorithimm the assigns <br>0 to female and 1 to male</p>

```python
le = preprocessing.LabelEncoder()
n_df["Gender_label"] = le.fit_transform(n_df["Gender"])
n_df.head()
```

<h3 style="font-weight:bold">2. One Hot Encoding</h3>

##### This type of encoding is used for x variate, It works by creating new columns for every varaiable in the x column. The algorithim then assign of 0 and 1

```python
#  To perform this type of encoding you can either use pandas or sklearn
df_one_hot = pd.get_dummies(n_df["Geography"])
df_one_hot
```

```python
# change all the categorical data in a dataframe to numerical data
df_one_hot = pd.get_dummies(df)
```
