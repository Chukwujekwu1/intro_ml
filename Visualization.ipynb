```python
stock_a = np.random.randint(low=30, high= 500, size=10)
stock_b = np.random.randint(low=30, high= 500, size=10)
stock_c = np.random.randint(low=30, high= 500, size=10)
```

```python
# initializing a line graph with diffrent data set
plt.figure(1)
plt.plot(stock_a, label = "Tesla")
plt.plot(stock_b, label = "Google")
plt.plot(stock_c, label = "Shell")
plt.legend()
plt.figure(2)
plt.hist(stock_a, label = "Tesla")
plt.hist(stock_b, label = "Google")
plt.hist(stock_c, label = "Shell")
plt.legend()
```
